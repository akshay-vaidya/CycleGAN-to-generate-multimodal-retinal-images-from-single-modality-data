```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("gen_loss.txt",delimiter="\n", header=None)
gen_loss = list(dataset[0])
dataset = pd.read_csv("discr_loss.txt",delimiter="\n", header=None)
dcr_loss = list(dataset[0])


x_list = range(len(gen_loss))
plt.plot(x_list, gen_loss, 'r', x_list, dcr_loss, 'b')
plt.show()
```
